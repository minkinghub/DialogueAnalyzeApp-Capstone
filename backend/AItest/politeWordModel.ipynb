```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification, pipeline

# TensorFlow 모델과 토크나이저를 불러옵니다.
# `from_pt=True`를 사용하여 PyTorch 가중치를 TensorFlow 모델로 변환합니다.
model = TFAutoModelForSequenceClassification.from_pretrained("j5ng/kcbert-formal-classifier", from_pt=True)
tokenizer = AutoTokenizer.from_pretrained('j5ng/kcbert-formal-classifier')

# pipeline 함수를 이용해 분류기를 생성합니다.
formal_classifier = pipeline(task="text-classification", model=model, tokenizer=tokenizer)
```

```python
# 분류기를 테스트합니다.
print(formal_classifier("나가뒤지십쇼"))
```
