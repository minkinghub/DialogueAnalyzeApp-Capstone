```python
# 11번가에서 제공하는 api를 이용한 검색 구현
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import os

# load .env
load_dotenv()

def api():
    key = os.getenv("11ST_API_KEY")
    keyword = "양파"
    url = "http://openapi.11st.co.kr/openapi/OpenApiService.tmall?key="+key+"&apiCode=ProductSearch&keyword="+keyword+"&sortCd=CP"
    response = requests.get(url)
    xmlRowdata = response.content.decode('cp949')
    soup = BeautifulSoup(xmlRowdata, 'html.parser')

    # 상품 정보를 저장할 리스트
    products = []

    # 각 상품별 정보 추출
    for product in soup.find_all("product"):
        product_info = {
            'productprice': product.productprice.get_text(),
            'productimage': product.productimage.get_text(),
            'productname': product.productname.get_text().strip(),
            'saleprice': int(product.saleprice.get_text()),
            'discount': product.discount.get_text(),
            'delivery': product.delivery.get_text(),
            'sellernick': product.sellernick.get_text(),
            'detailpageurl': product.detailpageurl.get_text(),
        }
        products.append(product_info)

    # saleprice를 기준으로 오름차순 정렬
    products = sorted(products, key=lambda x: x['saleprice'])

    # 결과 출력
    for product in products:
        print(f"상품명: {product['productname']}")
        print(f"원래 가격: {product['productprice']}")
        print(f"세일 가격: {product['saleprice']}")
        print(f"할인 금액: {product['discount']}")
        print(f"배달비: {product['delivery']}")
        print(f"판매자: {product['sellernick']}")
        print(f"상세 페이지 URL: {product['detailpageurl']}")
        print(f"상품 이미지 URL: {product['productimage']}")
        print("======================================")

if __name__ == "__main__":
    api()
```
